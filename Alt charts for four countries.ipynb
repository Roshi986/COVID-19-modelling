```python
grouped = pd.read_csv('COVID_july_grouped.csv', parse_dates=['Date'])
grouped
```

```python
grouped = grouped[[ 'Date', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active']]
```

```python
grouped
```

```python
#Brazil
brazil_all = grouped[grouped['Country/Region']=='Brazil']
```

```python
base = alt.Chart(brazil_all).mark_bar().encode(
    x='monthdate(Date):O',
).properties(
     width=500
)
```

```python
red = alt.value("#f54242")
green = alt.value('green')
purple = alt.value('purple')
orange = alt.value('orange')

base.encode(y='Confirmed').properties(title='Total confirmed cases in Brazil')|base.encode(y='Deaths', color=red).properties(title='Total deaths in Brazil') 
```

```python
base.encode(y='Recovered', color=orange).properties(title='Total recovered cases in Brazil')|base.encode(y='Active', color=purple).properties(title='Total active cases in Brazil')
```
